```python
import pandas as pd 
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
%matplotlib inline
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
new_folder = "./" #"./Results/fc1d_49c02v/"
```

```python
fName = ["x", "y", "dy"]
```

```python
res = {}
for fN in fName:
    res[fN] = np.genfromtxt(f"{new_folder}{fN}.csv", delimiter=',')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))

ax.plot(res["x"], res["y"], 'o-', lw=1, ms=3, label="y")

#ax.text(0.1, 0.1, 'text here', transform=ax.transAxes)
ax.set_xlabel(r'Current$\ (A) $', fontsize=15)
ax.set_ylabel(r'Mass Flow Rate$\ (g/s) $', fontsize=15)
# https://matplotlib.org/stable/gallery/text_labels_and_annotations/tex_demo.html
ax.tick_params(axis='both', which='major', labelsize=12)
#ax.grid('minor')
ax.legend( fontsize=13)
#plt.ylim([0.6, 1.12])

plt.savefig("test_All.png", transparent=True)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))

ax.plot(res["x"], res["dy"], 'o-', lw=1, ms=3, label="dy")

#ax.text(0.1, 0.1, 'text here', transform=ax.transAxes)
ax.set_xlabel(r'Current$\ (A) $', fontsize=15)
ax.set_ylabel(r'Mass Flow Rate$\ (g/s) $', fontsize=15)
# https://matplotlib.org/stable/gallery/text_labels_and_annotations/tex_demo.html
ax.tick_params(axis='both', which='major', labelsize=12)
#ax.grid('minor')
ax.legend( fontsize=15)
#plt.ylim([0.6, 1.12])

plt.savefig("test_All.png", transparent=True)
plt.show()
```
